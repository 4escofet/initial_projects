### Label encoder and One Hot encoder whith Scikit learn

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
# intantiate skl object
```

```python
df
```

```python
le.fit(df['sex']) # fitting on sex column from data set
print(le.classes_)
```

```python
df['sex_le'] = le.transform(df['sex']) # We encode on a new column using class object
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder2 = OneHotEncoder(sparse = False) #Intantiate OHE class object
```

```python
thales = df.thal.values.reshape(-1,1)
print(np.unique(thales))
```

```python
onehot_encoder2.fit(thales)
```

```python
onehot_encoder2.categories_
```

```python
thales_encoded = onehot_encoder2.transform(thales)
print(thales_encoded)
```

```python
df['fixed_defect_encoded'] = thales_encoded[:,0]
df['nromal_encoded'] = thales_encoded[:,1]
df['reversible_defect_encoded'] = thales_encoded[:,2]
df.head(100)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
# import class object to perform bins in age feature 
```

```python
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy = 'uniform')
# class object Intantiation 
```

```python
edades = df.age.values
print(edades.reshape(-1,1).shape)
```

```python
est.fit(edades.reshape(-1,1))
```

```python
est.bin_edges_ #bins ages
```

```python
df['bin_ages'] = est.transform(edades.reshape(-1,1))
# One line whith df['rangos_etarios_scikit'] = est.fit_transform(edades.reshape(-1,1))
```

```python
def plot_corre_heatmap(corr):
    '''
    Definimos una función para ayudarnos a graficar un heatmap de correlación
    '''
    plt.figure(figsize=(12,10))
    sns.heatmap(corr, cbar = True,  square = False, annot=True, fmt= '.2f'
                ,annot_kws={'size': 15},cmap= 'coolwarm')
    plt.xticks(rotation = 45)
    plt.yticks(rotation = 45)
    # Arreglamos un pequeño problema de visualización
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show()
```

```python
df2 = df.drop(['patient_id', 'thal', 'sex', 'age'], axis=1)
corr = df2.corr()
plot_corre_heatmap(corr)
```

```python
# Import class object for Decision Tree
from sklearn.tree import DecisionTreeClassifier

# Creamos un objeto arbol
tree = DecisionTreeClassifier(max_depth=3)
```

```python
df.columns
```

```python
X = df.drop(['patient_id', 'thal', 'sex', 'age','heart_disease_present'], axis=1)
y = df['heart_disease_present']
X.shape, y.shape
#df.drop(['Cabin','PassengerId','Name','Ticket','Embarked'], axis=1,inplace=True)
```

```python
tree.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score # SKL Metrics import

# Predict over training set 
y_pred = tree.predict(X)

# Compare whith real labels
accuracy_score(y_pred,y)
```

```python
tree2 = DecisionTreeClassifier()
tree2.fit(X, y)
y_pred = tree2.predict(X)
accuracy_score(y_pred,y)
```

### Whith DecisionTreeClassifier max_depth=3, the model had  84% of accuracy, and whith no deep limit reached 100% (seems overfitted)
