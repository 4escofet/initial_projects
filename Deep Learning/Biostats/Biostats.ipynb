```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df =pd.read_csv('biostats.csv', )
df['Name'] = df.iloc[:,0]
```

```python
df['Gender'] = df.iloc[:,1]
df.Gender
```

```python
df['age'] = df.iloc[:,2]
df['height'] = df.iloc[:,3]
df['weight'] = df.iloc[:,4]
df
```

```python
#df = df.iloc[:, [0,5,6,7,8]]
```

```python
df.columns
```

```python
df.drop(['age_int'],axis=1)
```

```python
#Encoding Sex Label
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Gender']) # fitting on sex column from data set
print(le.classes_)
```

```python
df['Gender'] = le.transform(df['Gender']) # We transform on a new column using class object
```

```python
#df.drop(['age_int'],axis=1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
# import class object to perform bins in age feature 
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy = 'uniform')
# class object Intantiation 
ages = df.age.values
print(ages.reshape(-1,1).shape)
est.fit(ages.reshape(-1,1))
est.bin_edges_ #bins ages
```

```python
df['bin_ages'] = est.transform(ages.reshape(-1,1))
```

```python
df.head()
```

```python
df.corr()
```

```python
X = df.drop(['Name','age','Gender'],axis=1)
y = df.Gender
```

```python
X.shape, y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X, y)
y_pred = tree.predict(X)
accuracy_score(y_pred,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_neighbors = 3
clf = KNeighborsClassifier(n_neighbors =n_neighbors )
clf.fit(X,y)
y_pred = clf.predict(X)
accuracy_score(y, y_pred)
```
